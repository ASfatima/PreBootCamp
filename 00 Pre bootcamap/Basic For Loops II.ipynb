```python
#Given a list, write a function that changes all positive numbers in the list to "big".
def changepoitivenum(list):
  for i in range(len(list)):
    if list[i]>0:
         list[i]='Big'
  return list

print(changepoitivenum([7,9,-6]))
```

```python
#Given a list of numbers, create a function to replace the last value with the number of positive values

def count_positives(list):
    count = 0
    for i in list:
        if i > 0:
            count += 1
    list[len(list)-1] = count
    return list

print(count_positives([-2,-91,0,17,-5,34]))
print(count_positives([6,-89,0,-3]))
```

```python
#Create a function that takes a list and returns the sum of all the values in the array.
def sumlist(list):
    sum = 0
    for i in list:
        sum += i
    return sum
print(sumlist([1,2,3,4]))
print(sumlist([5,6]))
```

```python
# Create a function that takes a list and returns the average of all the values.
def avergelist(list):
    sum=0
    count=0
    for i in list:
        sum += i
        count=len(list)   
    return (sum/count)
print(avergelist([5,4]))
print(avergelist([10,20,30,5]))
```

```python
# Create a function that takes a list of numbers and returns the minimum value in the list.
def minvalue(list):
    if len(list)<0:
        return False
    min_num = list[0]
    for i in list:
        if i<min_num:
            min_num = i
    return min_num
print("The minimum value =",minvalue([65,80,1,2,3,5]))
```

```python
# Create a function that takes a list and returns the maximum value in the array
def maxvalue(list):
    if len(list)<0:
        return False
    max_num = list[0]
    for i in list:
        if i>max_num:
            max_num = i
    return max_num
print("The maximum value =",maxvalue([65,80,1,2,3,5]))
```
