```python
#Create a function that accepts a number as an input.
def  countdown(x):
    nList = []
    for i in range(x, -1, -1):
        nList.append(i)
    return nList
print(countdown(6))
```

```python
#Create a function that accepts a list and returns the sum of the first value in the list plus the list's length
def sumfirstandlen(list):
    print("first number",list[0])
    print("length of list",len(list))
    return len(list)+list[0]
print("The sum of first number and length list = ",sumfirstandlen([2,4,6]))
```

```python
#Write a function that accepts two integers as parameters: size and value. 
def size_value(size,value):
    nlist=[]
    for i in range(size):
        nlist.append(value)
    return nlist
print(size_value(2,4))
print(size_value(6,10))
```
