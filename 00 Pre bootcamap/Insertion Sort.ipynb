```python
def insertion_sort(list_num):
    index_length = range(1, len(list_num))
    for i in index_length:
        value = list_num[i]

        while list_num[i-1] > value and i>0:
            list_num[i], list_num[i-1] = list_num[i-1], list_num[i]
            i = i -1

    return list_num



print(insertion_sort([6,5,3,1,8,7,2,4]))
```
