```python
def selection_sort(list_num):
  index_length= range(0,len(list_num)-1)

  for i in index_length:
    min_num=i

    for j in range(i+1 , len(list_num)):
      if list_num[j]< list_num[min_num]:
        min_num=j

        if min_num !=i:
          list_num[min_num],list_num[i]=list_num[i], list_num[min_num]

  return list_num

print(selection_sort([8,5,2,6,9,3,1,4,0,7]))


```
